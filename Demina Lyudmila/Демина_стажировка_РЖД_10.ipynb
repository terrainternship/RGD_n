##Скалируем остальные ('Серия', Период, 'Общий пробег', 'месяц') по мин-макс

```python
df_scaled = minmax.transform(df[cols_to_scale])
df_scaled = pd.DataFrame(df_scaled, columns = cols_to_scale)
```
